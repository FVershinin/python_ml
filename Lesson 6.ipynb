## Task 2

```python
n = np.linspace(1, 10, num=51)
```

```python
f = np.array([math.cos(it) for it in n])
```

```python
plt.plot(n, f)
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 5)
```
